### RAFI Indices

```python
import pandas as pd

df = pd.read_excel(f"US_indices/RAFI-US-Returns.xlsx", skiprows=9)
df = df.iloc[:, 1:]
df.head()
```

```python
date_col = df.iloc[2:,0]

indices = []
for col in df.columns[1:]:
    if col.__contains__("Index"):
        indices.append(col)
indices
```

```python
df = df.loc[2:, indices]
df.index = date_col
df.index.name = 'Date'
df.head()

for i in range(len(indices)):
    df[indices[i]].dropna().to_csv(f"{indices[i]}.csv", index=True)
```
